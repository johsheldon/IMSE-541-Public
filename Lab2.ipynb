```python
data1=pd.read_csv('Batch 1.csv')
data2=pd.read_csv('Batch 2.csv')
```

```python
pd.read_csv("BUSHING DIAMETER 2 BATCHES.csv")
def f_test(x,y):
  x=np.array(x)
  y=np.array(y)
  f=np.var(x,ddof=1)/np.var(y, ddof=1) # cal F test statistic
  dfn=x.size-1
  dfd=y.size-1
  p=1-stats.f.cdf(f,dfn, dfd) #find p-value of F test statistics
  return f, p
  
f_test(data1, data2)
```

```python
#two sample t-test 1: assume the variance is the same (default)
stats.ttest_ind(data1, data2, equal_var=True)
```

```python
#The P-Value is really high, higher than 0.05.  So we fail to reject the null hypothesis.  We can conclude that the data is not significantly different.
```
